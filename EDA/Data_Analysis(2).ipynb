```python
measure_unit = pd.read_csv("/content/drive/MyDrive/FoodData_Central_Clean/measure_unit.csv")
measure_unit.shape
```

```python
measure_unit.info()
```

```python
measure_unit.head(10)
```

```python
# Generate WordCloud from the 'name' column in the DataFrame
text = " ".join(measure_unit['name'].astype(str))  # Join all unit names into a single string

# Create the WordCloud
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    colormap='viridis'
).generate(text)

# Display the WordCloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")  # Turn off axis
plt.title("Word Cloud of Measuring Units", fontsize=16)
plt.show()
```
